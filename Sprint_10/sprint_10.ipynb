### Task 2:
##### a. Create a column that is the full name.
##### b. Create a column if the person was born in Spain or not.
##### c. Put the DNI as the index of the DataFrame (row names).
##### d. Replace the name of the columns Day of Birth, Month of Birth and Year of Birth with Day, Month and Year.
##### e. Replace H with Male, D with Female, A with Others and NC with a missing data (nan/null/na).
##### f. Show all the changes you have made in a single table.

```python
#a. Create a column full name 
df["Full Name"] = df["Nom"].astype(str) + " " + df["Cognoms"].astype(str)
df
```

```python
#b.
def spain(country):
    if country["País d'origen"]  == "Spain":
        return "Yes"
    else:
        return "Non"
df["Spanish"] = df.apply(spain, axis = 1)
df
```

```python
#c. DNI as index
df = df.set_index("DNI")
df
```

```python
#d. rename columns
df.rename(columns={'Dia de Naixement':'Day', 'Mes de Naixement':'Month', 'Any de Naixement':'Year'}, inplace=True )
df
```

```python
#b.
def replace(gender):
    if gender["Gènere"]  == "D":
        return "Woman"
    elif gender["Gènere"]  == "H":
        return "Man"
    elif gender["Gènere"]  == "A":
        return "Others"
    else:
        return "NaN"
df["Gènere"] = df.apply(replace, axis = 1)
df
```

### Task 3:
##### a. Merge the Children and Not Children columns into a single column, using the .apply() method and defining a function that solves the problem. The new column should be called "Children" and take the values ​​"Yes" or "No".

```python
#Creamos funcion con apply y generamos nueva columna "Children" en base a procesado de "Fills" y "No Fills"
def fills(user):
    if user["Fills"]  == 1.0:
        return "Yes"
    elif user["No Fills"] == 1.0:
        return "No"
    else:
        return "None"
df['Children'] = df.apply(fills, axis = 1)
df
```

### Task 4:
##### a. Create a summary table that allows you to see the average, median, minimum and maximum salary by Gender.
##### b. Sort the table based on average salary.

```python
#a. summary table
#Primero eliminamos simbolos de esta columna
df["Salari mensual"] = df["Salari mensual"].replace(r'[^\d.]', '', regex=True)
#Transformamos los valores de la columna en numeros
df["Salari mensual"] = pd.to_numeric(df["Salari mensual"], errors='coerce')
#df.groupby("Gènere")["Salari mensual"].describe() y bajamos a 2 decimales
salary = df.groupby("Gènere")["Salari mensual"].agg(['mean', 'median', 'min', 'max']).round(2)
#b. Ordenamos por la columna mean de mayor a menor
salary_desc = salary.sort_values(by="mean", ascending=False)
salary_desc
```

### Task 5:
##### a. Create a summary table with the average salary by gender (rows) and country of origin (columns).

##### b. Add the averages to the table margins.

##### c. (EXTRA): Apply conditional formatting to the table to see the highest values ​​in a more intense color

```python
#a y b.
pivot_table = df.pivot_table(
    values="Salari mensual", 
    index="Gènere", 
    columns="País d'origen", 
    aggfunc='mean', 
    margins=True, 
    margins_name='Total (Average)'
)
round_table = pivot_table.round(2)
round_table
```

```python
#c. 
styled_table = round_table.style.background_gradient(cmap='YlGnBu', axis=None).format(precision=2)
styled_table
```

### Task 6:
##### a. Create a new column that is the date of birth in Datetime format from the day, month, and year columns. 
##### b. Using this column, create a function that, given a date, calculates your current age as of today.
##### c. Use the function you just created to generate a new column in the DataFrame with the current age.

```python
#a. 
df["Date_of_birth"] = pd.to_datetime(df[["Year", "Month", "Day"]].rename(columns={
    "Year": "year",
    "Month": "month",
    "Day": "day"
}))
df
```

```python
#b. Generamos la funcion para calcular la edad en base a la fecha de nacimiento
from datetime import date

def calculateAge(birthDate):
    today = date.today()
    age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day))
    return age

#print(calculateAge(date((1997, 2, 3)), "Years"))

#c.  Aplicamos la función a cada fila de la columna
df["Age"] = df["Date_of_birth"].apply(calculateAge)
df
```

### Task 1:
##### a. Using the following DataFrame, attach the "Increment" column to the dataframe of the previous level.
##### b. Update the salary column based on the percentages attached. Do not manually modify the increments, write Python code for the necessary conversions.

```python
#a.:
df_increment = pd.DataFrame({"Grup":["Grup A","Grup B","Grup C", "Grup D" ] , "Increment":
["5%","3,5%","2%","8%"]})
df_increment
```

```python
new_df = pd.merge(
    left=df, 
    right=df_increment,
    how='left',
    left_on=["Grup Professional"],
    right_on=["Grup"],
)
#new_df.dtypes
# Assuming 'df' is your DataFrame and 'column_name' is the target
new_df['Increment'] = new_df['Increment'].str.replace('%', '', regex=False)
new_df['Increment'] = new_df['Increment'].str.replace(',', '.', regex=False)
new_df
```

```python
new_df['Increment'] = new_df['Increment'].astype(float)
new_df.dtypes
```

```python
#b. Calculamos un nuevo salario en base al % de incremento
new_df["New_salary"] = ((new_df["Salari mensual"] * new_df["Increment"])/100 + new_df["Salari mensual"]).round(2)
new_df
```

### Task 2:
##### a. Using a loop, export the data of each Grup Professional in 4 files (.xlsx or .csv format).
##### For example: "dades_GrupA.xlsx" , "dades_GrupB.xlsx" ...
##### b. Exports a 5è DataFrame in .xlsx or .csv format that contains how many workers there are for each Grup Professional, who is the most middle-aged.

```python
#a. We export each professional group in 4 formats
groups = new_df["Grup Professional"].unique()

for g in groups:
    # Filter the DataFrame for the specific group
    df_grup = new_df[new_df["Grup Professional"] == g]
    
    # Create the filename dynamically
    filename = f"dades_{g}.csv"
    
    # Export to Excel
    df_grup.to_csv(filename, index=False)

```
