# Task 1: IMC calculator



### We generate the function to calculate the IMC

```python
def imc_calculator(weight, height):
    """
    Calcula el indice de masa corporal.
    weight es el peso en kg (float)
    height es la altura en metro (float)
    """
    imc = weight/(height**2)
    imc = round(imc,2)

    """
    AHora clasificamos los resultados del imc en categorias
    """
    if imc < 18.5:
        category = "Under weight"
    elif 18.5 <= imc < 25:
        category = "Normal weight"
    elif 25 <= imc < 30:
        category = "Overweight"
    else:
        category = "Obesity"
    
    """
    Return both values imc anbd category
    """
    return imc, category
```


### We ask for input information, we validate introduced data: weight and height and run the function.

```python
# Data validation:
try: 
    weight = float(input("What your weight in Kg?:"))
    height = float(input("What is you height in Meter (ej. 1.65)?:"))
#Validacion de rangos de valores razonables:
    if weight <= 0 or weight > 600:
        print("Introduced values for weight are out of normal range")
    elif height <= 0.4 or height > 2.8:
        print("Introduced values for heaight are out of range")
    else:
        # Si los valores son correctos ejecutamos la funcion
        imc_value, category = imc_calculator(weight, height)
        print(f"Your IMC is {imc_value} so your physical condition is: {category}")
except ValueError:
    #Lo ejecutamos si los valores introducidos son numeros o simbolos
    print("The introduced values are not numeric. Please introduce valid numbers")
    
```


## Task 2: Temperature Conversor


### Conversor Function

```python
def temp_conversor (temp_type, temp_values):
    """
    1. Convierte cualquier temepratura ingresada primero a Celsius 
    """
    if temp_type == "C":
        celsius = temp_values
    elif temp_type == "F":
        celsius = (temp_values - 32) * 5/9
    elif temp_type == "K":
        celsius = temp_values - 273.15
    elif temp_type == "R":
        celsius = (temp_values - 491.67) * 5/9
    else:
        return None

    """
    2. Aplicamos la transformacion y las guardamos en un diccionario

    """
    conversions = {
        "Celsius": round(celsius, 2),
        "Fahrenheit": round((celsius * 9/5) + 32,2),
        "Kelvin": round(celsius + 273.15,2),
        "Rankine": round((celsius *9/5) + 491.67,2)
    } 

    return conversions
```


### Data validation


```python
# Input data validation:
try:
    temp_type = input("Introduce temperature unit: C, for Celsius, F for Fahrenheit, K for Kelvin or R: ").strip().upper()
    temp_values = float(input("Introduce a temperature value (ex: 24): "))

    #Kelvin can not be negative
    if temp_type == "K" and temp_values < 0:
        print("WARNING!: This value is under absolute zero")
    elif temp_type == "C" and temp_values < -273.15:
        print("WARNING!: This value is under absolute zero")
    else:
        conversion_results = temp_conversor (temp_type, temp_values)
        if conversion_results:
            print("Conversion results are saved in 'conversion_results' dictionary: ")
            print(conversion_results)
        else:
            print("Error in temperature unit")

except ValueError:
    print("Please, introduce a valid temperature value: ")
```


## Task 3: Word frequency counter and Average lenght of words calculator



### Words counter function


```python
def words_counter(text):
    characters_to_remove = ".,!?;:()\"'[]" #Esto remplazar por re.findall(, lower())
    clean_text = text.lower()

    for c in characters_to_remove:
        clean_text = clean_text.replace(c, "")

    # 1 Dividimos el texto en palabras
    strings = clean_text.split()

    # 2 Contamos la frecuancua de cada palabra  e iniciamos el contador de letras
    freq = {}
    sum_letters = 0
    #Tambien podria usar counter para contar la frequencia de las palabras y luego convertirlo en dict el resultado
    #Mean(map(len.dic{k}))
    for l in strings:
        freq[l] = freq.get(l,0) + 1
        sum_letters += len(l)

    # 3 Calculamos la media de letras por palabras
    avg = sum_letters / len(strings)

    return freq, round(avg,2)
```


### Data Input and Validation


```python
try:
    text = input("Introduce the text that you want to analyse: ") 
    # We call the function
    results_frequncies, average_letters = words_counter(text)

    #Now we check that results_frequencies is a dictionary:
    if isinstance(results_frequncies, dict):
        print("\n Results of frequencies: ")
        for word, number in results_frequncies.items():
            print(f"> '{word}': {number}")
            
        print(f"\n📏 The average lenght of a word in this text is: {average_letters}")
    else:
        print(results_frequncies)

except Exception as e:
    print(f"We have an error: {e}")
```


## TASK 4: Reverse dictionary to get a new survey organization



### Function to invert the order from key to values


```python
def invert_survey(original_survey):  
    inverted_survey = {}
    duplicates = {}
#Usar get en lugar del loop, una vez que creas el dicc vacio
    for key, value in original_survey.items():
        # Check if the value (which will be the new key) already exists
        if value in inverted_survey:
            # If it's the first time we find this duplicate, start a list with the existing key
            if value not in duplicates:
                duplicates[value] = [inverted_survey[value]]
            # Add the current key associated with this repeated value
            duplicates[value].append(key)
        else:
            # If it is unique so far, perform the swap
            inverted_survey[value] = key

    # Warning management
    if duplicates:
        print("WARNING: Duplicate values were found in the original survey.")
        for new_key, original_keys in duplicates.items():
            print(f"The new key '{new_key}' has these associated original keys: {original_keys}")
    
    return inverted_survey
```

```python
#Mirar salto de linea y replace >>>> Usar regex
def process_text_file(file_path):
    # Check if the file exists
    if not os.path.exists(file_path):
        return f"Error: The file '{file_path}' was not found."

    try:
        with open(file_path, 'r', encoding='latin-1', errors='replace') as f:
            text = f.read().lower()
        # Clean punctuation
        chars_to_remove = ".,!?;:()\"'[]\n"
        for c in chars_to_remove:
            text = text.replace(c, " ")

        words = text.split()
        
        # Sort all words alphabetically first to facilitate nested sorting
        words.sort()
        
        # Create the nested dictionary: {Initial: {word: frequency}}
        organized_data = {}

        for word in words:
            initial = word[0] # Get the first letter
            
            # If the initial letter is not a key yet, create a sub-dictionary
            if initial not in organized_data:
                organized_data[initial] = {}
            
            # Count the frequency within that specific initial
            organized_data[initial][word] = organized_data[initial].get(word, 0) + 1

        # Sort the main dictionary by keys (A to Z)
        final_sorted_data = {k: organized_data[k] for k in sorted(organized_data.keys())}
        
        return final_sorted_data

    except Exception as e:
        return f"An unexpected error occurred: {e}"
```

```python
# --- Execution ---
file_name = "/Users/tonzo/Desktop/IT_ACADEMY/Sprint_9/Extra_files/tu_me_quieres_blanca.txt"
results = process_text_file(file_name)


if isinstance(results, dict):
    print(f"Word Frequencies for '{file_name}' (Grouped by Initial):\n")
    for letter, word_dict in results.items():
        print(f"LETTER '{letter.upper()}':")
        for word, freq in word_dict.items():
            print(f"   - {word}: {freq}")
else:
    print(results)
```


## Task 2: transform to float

```python
def classify_elements(data_list):
    floats_list = []
    errors_list = []

    for item in data_list:
        # Si el elemento es una lista o tupla, procesamos su interior recursivamente
        if isinstance(item, (list, tuple)):
            sub_floats, sub_errors = classify_elements(item)
            floats_list.extend(sub_floats)
            errors_list.extend(sub_errors)
        else:
            try:
                # Intentamos la conversión
                number = float(item)
                floats_list.append(number)
            except (ValueError, TypeError): #Execption mirar como ponerlo aqui
                # Si falla (por ser texto o tipo no compatible), va a errores
                errors_list.append(item)
                
    return floats_list, errors_list
```

```python
# --- Datos del cliente ---
# Tupla  ('2', 1, 1.4, 'not-a-number')
#lista [1, 2, '3', '3.4']
#strings or non numeric one, seven
client_data = [
    '1.3', 'one', '1e10', 'seven', '3-1/2', 
    ('2', 1, 1.4, 'not-a-number'), 
    [1, 2, '3', '3.4']
]

# Ejecución
valid_floats, invalid_elements = classify_elements(client_data)

print("✅ Elements converted to float:")
print(valid_floats)

print("\n❌ Elements that could not be converted:")
print(invalid_elements)
```

## Task 1: Password generator

```python
def crear_contrasenya(longitud, majuscules=True, minuscules=True, numeros=True, signes=False):
    # 1. Definimos los conjuntos de caracteres posibles
    pool = ""
    if majuscules:
        pool += string.ascii_uppercase
    if minuscules:
        pool += string.ascii_lowercase
    if numeros:
        pool += string.digits
    if signes:
        pool += ".,-$;@#%&*"

    # 2. Validación: Si el usuario pone todo en False, no podemos generar nada
    if not pool:
        return "Error: Debes seleccionar al menos un tipo de carácter."

    # 3. Generamos la contraseña usando NumPy
    # Convertimos el pool en una lista para que NumPy pueda elegir
    pool_list = list(pool)
    #Aqui hay que asegurarse que hay uno de cada categoria
    # np.random.choice elige 'longitud' elementos al azar de la lista
    password_array = np.random.choice(pool_list, size=longitud)
    
    # Unimos el array de NumPy en un solo string
    password = "".join(password_array)
    pyperclip.copy(password) #aqui copiamos la contrasena en el portapapeles
    
    return password
```

```python
# 1. Definimos la ruta al archivo de stats
file_name = "historic_partits.txt"
```

```python
#En forma de funcion:
def analysis_stats(file_name):
    df = pd.read_csv(file_name, sep='\t', header=None)
    df.columns = ['Equip_local', 'Results', 'Equip_visit']

    # 2. Separar goles y convertir a números
    df[['goals_local', 'goals_visit']] = df['Results'].str.split('-', expand=True)
    df = df.astype({"goals_local": int, "goals_visit": int})

    # 3. Calcular puntos 
    df['punts_local'] = (df['goals_local'] > df['goals_visit']) * 3 + (df['goals_local'] == df['goals_visit']) * 1
    df['punts_visit'] = (df['goals_visit'] > df['goals_local']) * 3 + (df['goals_visit'] == df['goals_local']) * 1

    # 4.Creamos dos tablas y las juntamos despues
    # Tabla de estadísticas como local
    local_stats = df[['Equip_local', 'goals_local', 'goals_visit', 'punts_local']]
    local_stats.columns = ['Equip', 'Gols_Fets', 'Gols_Rebuts', 'Punts']

    # Tabla de estadísticas como visitante
    visit_stats = df[['Equip_visit', 'goals_visit', 'goals_local', 'punts_visit']]
    visit_stats.columns = ['Equip', 'Gols_Fets', 'Gols_Rebuts', 'Punts']

    # Concatenamos una debajo de otra y agrupamos por equipo
    classificacio = pd.concat([local_stats, visit_stats]).groupby('Equip').sum()

    # 5. Extraer los resultados
    sort_desc = classificacio.sort_values(by='Punts', ascending=False)
    total_goals = classificacio['Gols_Fets']
    equip_mes_golejador = classificacio['Gols_Fets'].idxmax()
    equip_mes_golejat = classificacio['Gols_Rebuts'].idxmax()

    return sort_desc, total_goals, equip_mes_golejador,equip_mes_golejat
```

```python
# Llamamos a la función y guardamos lo que devuelve en variables globales
sort_desc, total_goals, equip_mes_golejador, equip_mes_golejat = analysis_stats(file_name)

print("--- CLASSIFICACIÓ GLOBAL ---")
print(sort_desc)
print(f"\nNombre total de gols per equip:\n{total_goals}")
print(f"\nL'equip més golejador és: {equip_mes_golejador}")
print(f"L'equip més golejat és: {equip_mes_golejat}")
```
